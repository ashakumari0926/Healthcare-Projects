# STEP 1: Import Libraries

```python
# Import necessary libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP 2: Load Dataset

```python
# Import the dataset CSV file
# Load healthcare dataset CSV file
df = pd.read_csv("drug_dataset.csv")
```

```python
# Display first 5 rows to understand structure
df.head()
```

# STEP 3: Dataset Overview

```python
# Check shape of dataset (rows, columns)
print("Dataset Shape:", df.shape)
# Check column names and data types
df.info()
```

# STEP 4: Missing Values Check

```python
# Check missing values in each column
df.isnull().sum()
```

# STEP 5: Data Cleaning

```python
# Fill missing values in Allergy column with 'Unknown'
df['Allergy'] = df['Allergy'].fillna('Unknown')
```

```python
# Check missing values after update
df.isnull().sum()
```

```python
df['Allergy'].isnull().sum()
```

# STEP 8:  Drug Distribution

```python
# Set style for better visualization
sns.set(style="whitegrid")

# Plot most common drugs
plt.figure(figsize=(6,3))
df['Drug'].value_counts().plot(kind='bar', color='skyblue')

plt.title("Most Common Drugs Prescribed")
plt.xlabel("Drug Type")
plt.ylabel("Number of Patients")

plt.show()
```

# STEP 9: Age Distribution

```python
# Age distribution of patients
plt.figure(figsize=(6,3))

sns.histplot(df['Age'], bins=20, kde=True, color='Blue')

plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count of Patients")

plt.show()
```

# STEP 10: Blood Pressure Distribution

```python
# Blood Pressure category distribution
plt.figure(figsize=(6,3))

df['BP'].value_counts().plot(kind='bar', color='orange')

plt.title("Blood Pressure Distribution")
plt.xlabel("BP Level")
plt.ylabel("Number of Patients")

plt.show()
```

# STEP 11: Gender vs Drug Analysis

```python
# Relationship between Gender and Drug type
plt.figure(figsize=(6,3))

sns.countplot(x='Drug', hue='Sex', data=df)

plt.title("Drug Usage by Gender")
plt.xlabel("Drug Type")
plt.ylabel("Number of Patients")

plt.show()
```
